## Executive Summary - High-Level Overview

**Dataset Summary:**


```python
# High-level dataset summary
print("="*80)
print("DATASET SUMMARY")
print("="*80)

print(f"\nTotal Articles: {len(df):,}")
print(f"Date Range: {df['date'].min()} to {df['date'].max()}")
print(f"Unique Publishers: {df['publisher'].nunique():,}")
print(f"Unique Stocks: {df['stock'].nunique():,}")
print(f"Unique Headlines: {df['headline'].nunique():,}")

# Basic statistics
print(f"\n" + "="*80)
print("KEY STATISTICS")
print("="*80)
print(f"Mean Headline Length: {df['headline_length'].mean():.1f} characters")
print(f"Median Headline Length: {df['headline_length'].median():.1f} characters")
print(f"Mean Word Count: {df['headline_word_count'].mean():.1f} words")
print(f"Median Word Count: {df['headline_word_count'].median():.1f} words")

# Stock distribution
print(f"\n" + "="*80)
print("STOCK DISTRIBUTION")
print("="*80)
stock_dist = df['stock'].value_counts()
for stock, count in stock_dist.items():
    pct = count / len(df) * 100
    print(f"  {stock:6s}: {count:>8,} articles ({pct:>5.2f}%)")

# Publisher distribution (top 10)
print(f"\n" + "="*80)
print("TOP 10 PUBLISHERS")
print("="*80)
publisher_dist = df['publisher'].value_counts().head(10)
for pub, count in publisher_dist.items():
    pct = count / len(df) * 100
    print(f"  {pub[:50]:50s}: {count:>8,} articles ({pct:>5.2f}%)")
```

## Summary Visualizations

This section provides key visualizations that highlight the main findings from all analyses.


```python
# Create summary dashboard visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('EDA Summary Dashboard', fontsize=18, fontweight='bold', y=0.995)

# Plot 1: Stock distribution
ax1 = axes[0, 0]
stock_counts = df['stock'].value_counts()
bars1 = ax1.bar(range(len(stock_counts)), stock_counts.values, color='steelblue')
ax1.set_xticks(range(len(stock_counts)))
ax1.set_xticklabels(stock_counts.index, rotation=45, ha='right')
ax1.set_ylabel('Number of Articles', fontsize=11)
ax1.set_title('Articles per Stock', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')

# Add value labels
for i, count in enumerate(stock_counts.values):
    ax1.text(i, count, f"{count:,}", ha='center', va='bottom', fontsize=9)

# Plot 2: Top 10 publishers
ax2 = axes[0, 1]
top_10_pubs = df['publisher'].value_counts().head(10)
bars2 = ax2.barh(range(len(top_10_pubs)), top_10_pubs.values, color='coral')
ax2.set_yticks(range(len(top_10_pubs)))
ax2.set_yticklabels([pub[:30] for pub in top_10_pubs.index], fontsize=9)
ax2.set_xlabel('Number of Articles', fontsize=11)
ax2.set_title('Top 10 Publishers', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='x')
ax2.invert_yaxis()

# Add value labels
for i, count in enumerate(top_10_pubs.values):
    ax2.text(count, i, f" {count:,}", va='center', fontsize=8)

# Plot 3: Headline length distribution
ax3 = axes[1, 0]
ax3.hist(df['headline_length'], bins=50, edgecolor='black', alpha=0.7, color='mediumseagreen')
ax3.axvline(df['headline_length'].mean(), color='r', linestyle='--', linewidth=2, label=f'Mean ({df["headline_length"].mean():.1f})')
ax3.axvline(df['headline_length'].median(), color='b', linestyle='--', linewidth=2, label=f'Median ({df["headline_length"].median():.1f})')
ax3.set_xlabel('Headline Length (characters)', fontsize=11)
ax3.set_ylabel('Frequency', fontsize=11)
ax3.set_title('Headline Length Distribution', fontsize=12, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Plot 4: Publication timeline (if dates available)
ax4 = axes[1, 1]
if 'date' in df.columns and df['date'].notna().sum() > 0:
    df_valid_dates = df[df['date'].notna()].copy()
    df_valid_dates['year'] = df_valid_dates['date'].dt.year
    yearly_counts = df_valid_dates['year'].value_counts().sort_index()
    ax4.plot(yearly_counts.index, yearly_counts.values, marker='o', linewidth=2, markersize=8, color='gold')
    ax4.set_xlabel('Year', fontsize=11)
    ax4.set_ylabel('Number of Articles', fontsize=11)
    ax4.set_title('Publication Timeline', fontsize=12, fontweight='bold')
    ax4.grid(True, alpha=0.3)
else:
    ax4.text(0.5, 0.5, 'Date data not available', ha='center', va='center', transform=ax4.transAxes, fontsize=12)
    ax4.set_title('Publication Timeline', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('figures/eda_summary_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved: figures/eda_summary_dashboard.png")
```
